```python


import pandas as pd

data =  pd.read_csv('../../stock_day.csv')
data.head()
```

```python
#
p_change = data['p_change']
p_change.head()
```

```python
# 分区间方式1
# 将股票的价格分成10个区间
qcut =  pd.qcut(p_change,10)
print(qcut)
```

```python
# 查看每个区间出现的次数
print(qcut.value_counts())
```

```python
# 分区间方式2  自定义分区
# 自己指定分组区间
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
p_counts = pd.cut(p_change, bins)
print(p_counts.value_counts())
```

```python

"""分组数据转oneHot编码"""
dummies =  pd.get_dummies(p_counts)
dummies.head()
```
