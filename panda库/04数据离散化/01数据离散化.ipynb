```python
import pandas as pd

data = pd.read_csv(r'C:\Users\86132\Desktop\panda库\03保存与读取\stock_day.csv')
data
```

```python
#取出你要离散化的数据
p_change = data['p_change']
p_change
```

```python
#将股票的价格分成10个区间
qcut = pd.qcut(p_change,10)
qcut
```

```python
#查看每个区间出现的次数
print(qcut.value_counts())
```

```python
#自定义区间方式

bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
cut = pd.cut(p_change,bins)
print(cut.value_counts())
```

```python
#分组数据转one-hot编码
dummies = pd.get_dummies(cut)
dummies.head()
```
