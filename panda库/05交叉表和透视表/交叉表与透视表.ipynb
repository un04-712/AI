```python

import pandas as pd

data = pd.read_csv(r'C:\Users\86132\Desktop\panda库\03保存与读取\stock_day.csv')
data.head()
```

```python
data.index
```

```python
time =  pd.to_datetime(data.index)
```

```python
time.weekday
```

```python
data['week'] = time.weekday
data.head(10)
```

```python
import numpy as np

temp =  np.where(data['p_change']>0,1,0)
print(temp)
data['p_n'] = temp
```

```python
data.head(10)
```

```python
count =  pd.crosstab(data['week'],data['p_n'])
print(count)
```

```python
# 计算涨和跌的比例

row_totals = count.sum(axis=1)

rate = count.div(row_totals, axis=0)
print(rate)
```

```python
#透视表求解在weekday中涨跌的概率
data.pivot_table(index=['week'],values='p_n')
```
