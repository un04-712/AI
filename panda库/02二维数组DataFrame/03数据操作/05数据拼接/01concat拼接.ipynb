```python
"""

    pandas.concat(objs, *, axis=0, join='outer', ignore_index=False,
        keys=None, levels=None, names=None, verify_integrity=False,sort=False, copy=None)

        objs：参与连接的pandas对象的列表或元组
        axis：{0 index 1 columns}
        join
                outer表示并集
                inner表示交集
        ignore_index： 为True则重置索引
        keys：序列，用于构造分层索引的键
        levels：序列列表  用于构造分成索引特定的级别
        verify_integrity：为True则检查新连接的轴是否包含重复索引，如果是，则报错
        sort：是否对结果按照列名进行排序 默认值 False
        copy：为none表示自动选择   True：拷贝底层数据  False：尽量不拷贝
"""
import pandas as pd
# 多个series对象拼接
ser1=pd.Series([1,2,3],index=list('ABC'))
ser2=pd.Series([4,5,6],index=list('DEF'))
print(ser1)
print(ser2)
print(pd.concat([ser1,ser2]))
```

```python
# 两个df对象拼接，默认找相同的列索引进行合并  行方向拼接 行索引重复

df1 = pd.DataFrame(
    {
        'A' : ['A0','A1','A2','A3'],
        'B' : ['B0','B1','B2','B3'],
        'C' : ['C0','C1','C2','C3'],
        'D' : ['D0','D1','D2','D3'],
    }
)
df2 = pd.DataFrame(
    {
        'A' : ['A4','A5','A6','A7'],
        'B' : ['B4','B5','B6','B7'],
        'C' : ['C4','C5','C6','C7'],
        'D' : ['D4','D5','D6','D7'],
        'E' : ['E4','E5','E6','E7'],
    }
)
result = pd.concat([df1,df2])
print(result)
print(result.at[0,'A'])
```

```python

# 两个df对象拼接，列的方式拼接  列索引重复
result = pd.concat([df1,df2],axis=1)
print(result)
```

```python

# 解决索引重复的问题 加 ignore_index 属性  重置索引
result = pd.concat([df1,df2],axis=1,ignore_index=True)
print(result)
```

```python
# 解决索引重复的问题 加 keys属性
result = pd.concat([df1,df2],axis=1,keys=['df1','df2'])
print(result)
```

```python
result = pd.concat([df1,df2],axis=1,keys=['df1','df2'])
print("获取df1块中的A列中的第一行",result['df1']['A'][0])
```

```python
# 两个df对象拼接  join 可以指定外连接 ： 会保留所有的列，如果某一个列某个df中有缺失（没有这一列），则会用nan填充。
result = pd.concat([df1,df2],axis=0,join='outer',ignore_index=True)
print(result)
```

```python
# 两个df对象拼接  join 可以指定内连接 ： 内连接会过滤掉那些只在一个 DataFrame 中存在而在另一个中没有的列（或行）标签。
result = pd.concat([df1,df2],axis=0,join='inner',ignore_index=True)
print(result)
```
