```python
"""
    按照索引顺序进行排序
    Series.sort_index(axis=0, level=None, ascending=True,inplace=False,
     kind='quicksort', na_position='last',sort_remaining=True, ignore_index=False, key=None)

    level：如果索引是多级索引（也称为层次化索引或MultiIndex），则可以指定先要排序的级别
    ascending：True为升序，False为降序
    inplace：是否在原始series中操作
    kind ： 排序算法
        quicksort：速度最快的，性能平均 但不稳定，最坏的情况下性能较差
        mergesort： 稳定的排序算法，性能较好，但是空间复杂度高
        heapsort: 空间效率较高，但不稳定
        stable： 稳定的排序算法，由mergesort和timesort实现
    na_position: 决定NaN值的放置位置  first放在首位  last放在尾部
    sort_remaining：如果为True，在level排序完成之后，会在排序基础上对剩下的级别元素还会排序
   ignore_index： 为True则重置索引
   key：传入函数地址   如果指定，则这个函数将在排序之前应用于每个值，并且排序将基于这些函数返回的值
"""

import pandas as pd
import numpy as np
'''
    基本使用方法
'''
s = pd.Series([10,11,12,13,14,15,16,np.nan,43],index=[1,2,4,3,9,np.nan,5,23,56])

print(s.sort_index())
```

```python
"""
    多级索引
"""
import pandas as pd
import numpy as np
arrays = [np.array(['qux', 'qux', 'foo', 'foo',
                    'baz', 'baz', 'bar', 'bar']),
          np.array(['two', 'one', 'two', 'one',
                    'two', 'one', 'two', 'one'])]

s = pd.Series([1, 2, 3, 4, 5, 6, 7, 8], index=arrays)
print(s)


```

```python
res = s.sort_index(level=1,sort_remaining=True)  # 先排第2级再排第1级 按升序进行排序
print(res)
```

```python
# dataframe进行排序
"""
    按照索引顺序进行排序
    DataFrame.sort_index(axis=0, level=None, ascending=True,inplace=False,
     kind='quicksort', na_position='last',sort_remaining=True, ignore_index=False, key=None)
    axis: 0 index  1 columns
    level：如果索引是多级索引（也称为层次化索引或MultiIndex），则可以指定先要排序的级别
    ascending：True为升序，False为降序
    inplace：是否在原始series中操作
    kind ： 排序算法
        quicksort：速度最快的，性能平均 但不稳定，最坏的情况下性能较差
        mergesort： 稳定的排序算法，性能较好，但是空间复杂度高
        heapsort: 空间效率较高，但不稳定
        stable： 稳定的排序算法，由mergesort和timesort实现
    na_position: 决定NaN值的放置位置  first放在首位  last放在尾部
    sort_remaining：如果为True，在level排序完成之后，会在排序基础上对剩下的级别元素还会排序
   ignore_index： 为True则重置索引
   key：传入函数地址   如果指定，则这个函数将在排序之前应用于每个值，并且排序将基于这些函数返回的值
"""


import pandas as pd
import numpy as np

arrays = [np.array(['qux', 'qux', 'foo', 'foo']),
          np.array(['two', 'one', 'two', 'one'])]

df = pd.DataFrame({'A':[1,2,3,4],'B':[4,3,2,1]},index=arrays)
print(df)
```

```python
res = df.sort_index(level=0)  # 先排第2级再排第1级 按升序进行排序
print(res)
```
