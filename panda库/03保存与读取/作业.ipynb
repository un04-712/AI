```python
#展示美国的星巴克数量和中国的数量进行对比，展示中国每个省份星巴克的数量的情况。可以额外增加统计功能。数据文件为directory.csv

import sqlite3

import pandas as pd

df = pd.read_csv(r"C:\Users\86132\Desktop\panda库\03保存与读取\directory.csv", encoding='utf-8')
data = df[df['Country']=='CN']
data
```

```python
CN_Star_num = df[df['Country']=='CN']["Country"].count()
print('星巴克在中国的数量是:',CN_Star_num)
```

```python
US_Star_num = df[df['Country']=='US']["Country"].count()
print('星巴克在美国的数量是:',US_Star_num)
```

```python
from matplotlib import pyplot as plt
labels = ['CN','US']
size = [CN_Star_num,US_Star_num]
plt.pie(size,
        labels=labels,
        autopct='%1.1f%%',
        textprops=dict(color='white')
        )
plt.title('中美Starbucks门店数量对比')
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用雅黑字体
plt.rcParams['axes.unicode_minus'] = False  # 显示负号
plt.legend()

plt.show()
```

```python
china_province_count = data['State/Province'].value_counts()
print(china_province_count.sort_index(ascending=True))
```

```python
plt.bar(china_province_count.index, china_province_count.values)
for i, value in enumerate(china_province_count.values):
    plt.text(i, value + 0.5, str(value), ha='center', fontsize=10)
plt.title('各省份门店数量分布')
plt.xlabel('省份')
plt.ylabel('门店数量')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
province_ratio = china_province_count / CN_Star_num
labels = china_province_count
size = province_ratio
plt.figure(figsize=(60,60))  

plt.pie(size,
        autopct='%1.1f%%',
        textprops=dict(color='white'),
        labels=labels
        )

plt.show()
plt.show()
```
