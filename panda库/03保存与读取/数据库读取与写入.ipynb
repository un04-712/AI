```python
import sqlite3

import pandas as pd
# 根据products表计算出总商品数量，总库存数量，平均库存数量，库存商品总价值，所用技术点不限
#数据库连接
sql = sqlite3.connect(r'C:\Users\86132\Desktop\numpy库\test.db')
#连接数据库
df1 = pd.read_sql('select * from products', con =  sql)

total_products_nums = df1['id'].count()
total_stock_nums = df1['quantity'].sum()
avg_stock_nums = total_stock_nums / total_products_nums
total_stock_num_values = (df1['price'] * df1['quantity']).sum()

print('总商品数量',total_products_nums)
print('总库存数量',total_stock_nums)
print('平均库存数量',avg_stock_nums)
print('库存商品总价值',total_stock_num_values)
```
