```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

x,y = np.meshgrid(x, y) #生成网格点坐标矩阵的函数,常用于场景二维码或者多维坐标网格
plt.scatter(x, y)
#计算Z轴的所有值

```

```python
z = np.sqrt(x**2 + y**2)    

plt.contourf(x,y,z)
plt.colorbar()
```

```python
from matplotlib.gridspec import GridSpec
import matplotlib.pyplot as plt
import numpy as np
import mplcursors
%matplotlib notebook
plt.rcParams['font.sans-serif'] = ['SimHei']  # 使用雅黑字体
plt.rcParams['axes.unicode_minus'] = False 

x1 = np.linspace(-10, 10, 100)
y1 = np.linspace(-10, 10, 100)

x2 = np.linspace(-50, 50, 300)
y2 = np.linspace(-50, 50, 300)
x1,y1 = np.meshgrid(x1, y1) #生成网格点坐标矩阵的函数,常用于场景二维码或者多维坐标网格
x2,y2 = np.meshgrid(x2, y2)
# 设置支持中文的字体
z1 = np.sqrt(x1**2 + y1**2)  
z2 = np.sqrt(x2**2 + y2**2)

fig = plt.figure()
gs = GridSpec(3,3)
ax1 = fig.add_subplot(gs[1,1],projection='3d',)
ax2 = fig.add_subplot(gs[2,2:4],projection='3d')

surf1 = ax1.plot_surface(x1,y1,z1)
surf2 = ax2.plot_surface(x2,y2,z2)


fig.colorbar(surf1,label='z1')
fig.colorbar(surf2,label='z2')


```
